* We can see from the above heatmap that `Sales` (label) has positive correlation with `TV`, `radio` and `newspaper` column with correlation value 0.78, 0.58 and 0.23 respectively.

* The independent variable `radio` and `TV`  have 0.055 correlation value which is quite small. The assumption of linear regression suggests that the independent variable should not be correlated. If multi-colinarity. In such a case we should drop one column.

In our case we will drop radio and newspaper as TV has the highest correlation value with sales.

```python
# selecting feature
x1 = df[['TV']]
y1 = df.sales
```

Train test split

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(x1,y1, test_size = 0.2, random_state = 42)
plt.scatter(x=X_train1, y = y_train1, label = 'Train Data')
plt.scatter(x=X_test1, y= y_test1, label = 'Test Data')
plt.legend()
plt.show()
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train1,y_train1)
y_pred1 = model1.predict(X_test1)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, median_absolute_error

MAE = mean_absolute_error(y_true = y_test1, y_pred = y_pred1)
MSE = mean_squared_error(y_true=y_test1, y_pred=y_pred1)
R2S = r2_score(y_true = y_test1, y_pred = y_pred1)
MEAE = median_absolute_error(y_true = y_test1, y_pred = y_pred1)
c = model1.intercept_
m = model1.coef_


print(f"Mean Absolute Error (MAE): {MAE:.2f}")
print(f"Mean Squared Error (MSE): {MSE:.2f}")
print(f"R-squared (R2): {R2S:.2f}")
print(f"Median Absolute Error (MedAE): {MEAE:.2f}")
print(f"Intercept (c): {c:.2f}")
```

```python
X = df.drop('sales', axis=1)
y = df[["sales"]]
```

Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=46)
plt.scatter(x=X_train1, y = y_train1, label = 'Train Data')
plt.scatter(x=X_test1, y= y_test1, label = 'Test Data')
plt.legend()
plt.show()
```

Liner Regression Model

```python
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model_linear.predict(X_test)
```

```python
MAE = mean_absolute_error(y_true = y_test, y_pred = y_pred)
MSE = mean_squared_error(y_true=y_test, y_pred=y_pred)
R2S = r2_score(y_true = y_test, y_pred = y_pred)
MEAE = median_absolute_error(y_true = y_test, y_pred = y_pred)
c = model1.intercept_
m = model1.coef_


print(f"Mean Absolute Error (MAE): {MAE:.2f}")
print(f"Mean Squared Error (MSE): {MSE:.2f}")
print(f"R-squared (R2): {R2S:.2f}")
print(f"Median Absolute Error (MedAE): {MEAE:.2f}")
print(f"Intercept (c): {c:.2f}")
```
